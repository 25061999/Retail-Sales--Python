```python
#Double check if the date is actually the date 
date_column=pd.DatetimeIndex(rs['Date'])
rs['Year']=date_column.year
rs['Month']=date_column.month
rs['Day']=date_column.day
```

```python
display(rs)
```

```python
#We got the sum of the revenue per month 
rs.groupby('Month')['Total Amount'].sum()
```

```python
display(rs)
```

```python

sales_by_category = rs.groupby('Product Category')['Total Amount'].sum()
plt.figure()
sales_by_category.plot(kind='bar')
plt.title('Total Sales by Category')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.show()
```

```python
# Sales Trend Over Time
rs['Date'] = pd.to_datetime(rs['Date'])
sales_trend = rs.groupby('Date')['Total Amount'].sum()
plt.figure()
sales_trend.plot()
plt.title('Sales Trend Over Time')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.show()
```

```python
# Seaborn - Sales by Month
plt.figure()
sns.barplot(data=rs, x='Month', y='Total Amount')
plt.title('Sales by Month')
plt.show()
```

```python
plt.figure()
plt.hist(rs['Total Amount'], bins=10)
plt.title('Distribution of Sales Amount')
plt.xlabel('Sales Amount')
plt.ylabel('Frequency')
plt.show()
```

```python
sales_by_category = rs.groupby('Product Category')['Total Amount'].sum()

plt.figure()
sales_by_category.plot(kind='pie', autopct='%1.1f%%')
plt.title('Sales Contribution by Category')
plt.ylabel('')
plt.show()
```

```python
sales_trend = rs.groupby('Date')['Total Amount'].sum()

plt.figure(figsize=(10,5))
plt.plot(sales_trend, marker='o')
plt.title('Sales Trend with Markers')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.show()
```

```python
# Sales Trend Over Time -Changing Color & Plot Size
rs['Date'] = pd.to_datetime(rs['Date'])
sales_trend = rs.groupby('Date')['Total Amount'].sum()
plt.figure(figsize=(12,5))
sales_trend.plot(color='green')
plt.title('Sales Trend Over Time')
plt.xlabel('Date')
plt.ylabel('Sales')

plt.show()
```
